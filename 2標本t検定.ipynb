## 必要なモジュールのインポート

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 正規分布（ガウス分布、ガウシアン）を生成

```python
mu = 100  # 平均
sigma = 5  # 標準偏差
a = np.random.normal(mu, sigma, 1000)  # 正規分布を生成
b = np.random.normal(mu+5, sigma, 1500)  # 正規分布を生成
```

## 生成した正規分布のヒストグラム（度数分布）を表示

```python
import pandas as pd
import seaborn as sns

plt.figure()
sns.histplot(a, color='dodgerblue', label='a')
sns.histplot(b, color='deeppink', label='b')
plt.legend()  # 判例を表示
plt.show()
```

## 2標本t検定を計算

```python
n1 = len(a)
n2 = len(b)

s1_2 = a.var(ddof=1)  # aの不偏分散
s2_2 = b.var(ddof=1)  # bの不偏分散

s_2 = ((n1-1)*s1_2 + (n2-1)*s2_2) / (n1+n2-2)

t = (a.mean() - b.mean()) / s_2**0.5 / (1/n1 + 1/n2)**0.5
```

## 2標本t検定の値を表示

```python
print(f'2標本t検定の値 = {t}')
```

## 2標本t検定を計算する関数を定義

```python
def twosample_ttest(x, y):
    n1 = len(a)
    n2 = len(b)
    
    s1_2 = a.var(ddof=1)  # aの不偏分散
    s2_2 = b.var(ddof=1)  # bの不偏分散
    
    s_2 = ((n1-1)*s1_2 + (n2-1)*s2_2) / (n1+n2-2)
    
    t = (a.mean() - b.mean()) / s_2**0.5 / (1/n1 + 1/n2)**0.5
    return t
```

## 平均の差を変更しながら検定統計量とヒストグラムを比較

```python
diffs = np.arange(10)

for diff in diffs:
    a = np.random.normal(mu, sigma, 500)
    b = np.random.normal(mu+diff, sigma, 500)

    t = twosample_ttest(a, b)
    print(f'平均の差 = {diff}, 2標本t検定の値 = {twosample_ttest(a, b)}')

    plt.figure()
    sns.histplot(a, color='dodgerblue', label='a')
    sns.histplot(b, color='deeppink', label='b')
    plt.legend()  # 判例を表示
    plt.show()
    
    print('\n\n')
```
